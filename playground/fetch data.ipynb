# 1. Get team season schedule

```python
schedule = SeasonSchedule('2022').get_schedule()
```

```python
team_ext_id = 'royal'
games_ids = schedule[(schedule['homeTeamID'] == team_ext_id) | (schedule['awayTeamID'] == team_ext_id)]['gameID'].to_list()
```

```python

dfs = []
for game_id in games_ids:
    data = compute_team_throwing_selection(game_id, team_ext_id)
    dfs.append(data)
```

```python
def get_throw(x, y):
    throw, side, x, y, angle = get_throw_type(0, 0, x, y)
    return throw

season_data = pd.concat(dfs)
season_data['throw_outcome'] = season_data['throw_type'].apply(lambda x: 'Completed' if x != 'Throwaway' else 'Throwaway')
season_data['throw_type'] = season_data.apply(lambda row: get_throw(row.x, row.y), axis =1)
season_data.head()
```

```python
season_data.to_csv('data/royal_2022.csv')
```
